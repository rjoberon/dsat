```python
import pandas as pd

"""
width x height = pixels
1024 x 2479 = 2538496 
bytes: 2538456 â†’ 40 pixels more

rows that roughly divide parts (last row of previous part):
1. 0 
2. 553
3. 1050
4. 1508
5. 1574
"""

with open("../un2.dat", "rb") as f:
    first = f.read(1024 * 553)
```

## Part 2.1

```python
from struct import unpack
import pandas as pd

with open("../un2.dat", "rb") as f:
    vals = []
    pos = 0
    while ((data := f.read(1)) and pos < 1024*553):
        vals.append(int.from_bytes(data, byteorder="little", signed=False))
        pos += 1

df = pd.DataFrame(vals, columns=["ints"])
```

### Autocorrelation

```python
import statsmodels.tsa.stattools as smtsa
import numpy as np
import matplotlib.pyplot as plt

acf = smtsa.acf(df.ints, nlags=50, adjusted=False, fft=False)

lags = np.arange(len(acf))
plt.rcParams['figure.figsize'] = (10, 5)

plt.vlines([6, 10, 16], -0.2, 0.8, color="lightgrey")
plt.plot(lags[1:], acf[1:])
plt.xlabel("bytes")
plt.xlim(xmin=0)
plt.ylabel("correlation")
plt.show()
```

### Map

```python
from struct import unpack
import pandas as pd

bytelen = 8

with open("../un2.dat", "rb") as f:
    vals = []
    pos = 0
    while ((data := f.read(bytelen)) and pos < 1024*554 + 464):
        vals.append([
            unpack("<I", data[0:4])[0],
            unpack("<I", data[4:8])[0]
        ])
        pos += bytelen

df = pd.DataFrame(vals, columns=["x", "y"])
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 10)
plt.rcParams['figure.dpi'] = 140
plt.plot(df.x, df.y, 'o', markersize=0.1)
plt.gca().invert_yaxis()
plt.gca().set_aspect('equal')
plt.show()
```
